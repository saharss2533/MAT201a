```python
import scipy.ndimage as ndimage
```

```python
subplot(121)
min_filter_sea = ndimage.minimum_filter(c, (40,10))
imshow(min_filter_sea)

subplot(122)
max_filter_sea = ndimage.maximum_filter(c, (500,30))
imshow(max_filter_sea)
```
