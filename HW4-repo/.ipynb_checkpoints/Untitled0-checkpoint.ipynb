```python
head2 = where(head > 40, 1, -1)
# imshow(head2)
imshow(head2, cmap=cm.gray, interpolation='nearest')
```

```python
from scipy.signal import correlate2d
cc = correlate2d(img, head2)
imshow(cc)
colorbar()
gcf().set_figheight(8)
```

```python
imshow(cc, cmap=cm.gray)
colorbar()
gcf().set_figheight(8)
```

```python
cc.shape
```

```python
avg=0
for x in range(765):
  for y in range(520):
        avg= avg+cc[x][y]
#     print (img[x][y])

avg=avg/(765*520)
print(avg)
```

```python
# detecting people by finding their heads, mostly those facing forward, 
# since our head model is a copy and pastee of man's head looking frw
# It is not super accurate because there are too many details in the picture
# that resembles human head, but overally it hedets mostly where people
# are located
subplot(121)
imshow(where(cc > 60, 1, 0), interpolation='nearest', cmap=cm.gray)
colorbar()
# gcf().set_figheight(8)
subplot(122)
imgOrig = imread('tehran-green.jpg')
imshow(imgOrig)
```

```python
# the red dots(the darker the closer to the other model) are the max values 
# of image that are the heads detected in the image. And the Blues are the
# other features of the image such as the space in between people, their 
# bodies and the fabric in the middle that does not resemble the fabric

from scipy.ndimage.filters import maximum_filter
subplot(121)
imshow(maximum_filter(cc, (10,10)))
subplot(122)
imshow(imgOrig)
```

```python
from scipy.signal import correlate2d
cc2 = correlate2d(img, paper2)
imshow(cc2)
colorbar()
gcf().set_figheight(8)
```

```python
#  getting the average to have an idea what the mean pixel value is
avg3=0
for x in range(cc2.shape[0]):
  for y in range(cc2.shape[1]):
        avg3= avg3+cc2[x][y]
#     print (img[x][y])

avg3=avg3/(cc2.shape[0]*cc2.shape[1])
print(avg3)
```

```python
# It detects the papers in people's hands.
# It detects all the papers in the back, since my model of paper is from 
# that part of the image and works for that dimension the best
# However, it has some false positives, which are some of the people's shirt 
# in the front that their shape is similar to the model of a paper.
# the max values in the matrix are the white pixel here, as described above, which 
# are the papers, and materials resembling a while paper that is detected
# in the image.
subplot(121)
imshow(where(cc > 200, 1, 0), interpolation='nearest', cmap=cm.gray)
colorbar()
subplot(122)
imgOrig = imread('tehran-green.jpg')
imshow(imgOrig)
```

```python
# The highest values of the matrix are the red colored pixels in the pic
# And it makes sense since it finds the original paper on top left corner
# (100<y<200, 0<x<100) and almost most of the papers that people are holding
# mainly the ones with less features on the paper and with similar size as
# or paper model.
# its false positives are the white shirts of some people and the fabric in 
# the middle, since they resemble a piece of paper.
# the lowest values are the blue pixels that are the heads and hands and other 
# parts of the image that does not resemble the piece of paper. 
from scipy.ndimage.filters import maximum_filter
subplot(121)
imshow(maximum_filter(cc2, (8,8)))
# gcf().set_figheight(8)
subplot(122)
imshow(imgOrig)
```
